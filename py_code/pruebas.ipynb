```python
import stocksOps as so
import matplotlib.pyplot as plt
```

```python
# Importing data from yfinance
data_nvidia = so.downloadData("NVDA",   "2021-06-19", "2022-06-19")
data_amd =    so.downloadData("AMD",    "2021-06-19", "2022-06-19")
```

```python
stats_nvidia = so.downloadTickerStatistics("NVDA")
print(stats_nvidia.info["returnOnEquity"])
print(stats_nvidia.financials)
```

```python
# Calculating closing returns
so.closingReturns(data_nvidia)
so.closingReturns(data_amd)

data_nvidia.head()
```

```python
data_amd.head()
```

```python
# Calculation of average returns
nvidiaAverageReturns    = so.averageReturns(data_nvidia)
amdAverageReturns       = so.averageReturns(data_amd)
print(f" avg returns nvidia: {nvidiaAverageReturns}")
print("{:.3%}".format(round(nvidiaAverageReturns, 4)))
print("{:.3%}".format(round(amdAverageReturns, 4)))
```

```python
# Calculation of variance returns
nvidiaVarianceReturns   = so.varianceReturns(data_nvidia, nvidiaAverageReturns)
amdVarianceReturns      = so.varianceReturns(data_amd, amdAverageReturns)
print(f"nvidia variance returns : {nvidiaVarianceReturns}")
print(f"amd variance returns: {amdVarianceReturns}")
```

```python
# Calculating the standard deviation of each stock dataset
nvidiaStandardDeviation = so.stdDeviation(nvidiaVarianceReturns)
amdStandardDeviation    = so.stdDeviation(amdVarianceReturns)
print(f"amd standard deviation : {amdStandardDeviation}")
print(f"nvidia standard deviation : {nvidiaStandardDeviation}")
```

```python
# Calculation of amd and nvidia covariance
amdNvidiaCovariance = so.covariance(data_nvidia, data_amd, nvidiaAverageReturns, amdAverageReturns)
print(f"covariance between amd and nvidia stock: {amdNvidiaCovariance}")
```

```python
# Calculation of amd and nvidia correlation
amdNvidiaCorrelation = so.correlation(amdNvidiaCovariance, nvidiaVarianceReturns, amdVarianceReturns)
print(f"correlation between amd and nvidia stock: {amdNvidiaCorrelation}")
```
